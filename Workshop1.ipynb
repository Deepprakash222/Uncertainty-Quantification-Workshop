# Probability Space: Basic Concepts

A probability space is the mathematical foundation for probability theory. It consists of three components:

[ ($\Omega$, $\mathcal{F}$, $P$) ]

where:

* ($\Omega$): Sample Space
* ($\mathcal{F}$): Event Space (($\sigma$)-algebra)
* ($P$): Probability Measure

---

## 1. Sample Space ($\Omega$)

The **sample space** is the set of all possible outcomes of a random experiment.

### Examples:

* Coin toss:
  $\Omega = \{H, T\}$ 
* Dice roll:
  $\Omega = \{1, 2, 3, 4, 5, 6\}$ 
* Continuous example (choosing a real number between 0 and 1):
  $\Omega = [0,1] $

---

## 2. Event Space ($\mathcal{F}$)

The **event space** is a collection of subsets of ($\Omega$). Each subset is called an **event**.

$\mathcal{F}$ must satisfy the properties of a **($\sigma$)-algebra**:

1. **Contains the sample space**:
   $\Omega \in \mathcal{F}$ 

2. **Closed under complement**:
   If $A \in \mathcal{F}$, then
   $A^c \in \mathcal{F}$ 

3. **Closed under countable unions**:
   If $A_1, A_2, A_3, \dots \in \mathcal{F}$, then
   $\bigcup_{i=1}^{\infty} A_i \in \mathcal{F}$ 



### Examples:

* Finite case (dice):
  Often, $\mathcal{F}= \mathcal{P}(\Omega) = \{\emptyset, H, T, \{H,T\}\}$ .

* Continuous case:
  ($\mathcal{F}$) is typically the **Borel ($\sigma$)-algebra** on ([0,1]).

---

## 3. Probability Measure ($P$)

The **probability measure** assigns a number to each event in $\mathcal{F}$.

$P: \mathcal{F} \rightarrow [0,1]$ 

It must satisfy the following axioms:

### 1. Non-negativity:

$P(A) \geq 0 \quad \forall A \in \mathcal{F}$ 

### 2. Normalization:

$P(\Omega) = 1 $

### 3. Countable Additivity (($\sigma$)-additivity):

If $A_1, A_2, A_3, \dots $ are disjoint events, then
$P\left(\bigcup_{i=1}^{\infty} A_i\right) = \sum_{i=1}^{\infty} P(A_i)$ 

---

### Putting It All Together

A **probability space** is the triple:

[ ($\Omega$, $\mathcal{F}$, $P$) ]

* ($\Omega$): What can happen
* ($\mathcal{F}$): What events we can measure
* ($P$): How likely those events are

---

### Example: Fair Coin Toss

* Sample space:
  $$\Omega = \{H, T\} $$

* Event space:
  $\mathcal{F} = \{ \emptyset, {H}, {T}, \Omega \} $

* Probability measure:
  
  $P(\emptyset) = 0.0, \quad P({H}) = 0.5, \quad P({T}) = 0.5, P(\Omega) = 1.0$
  

---

### Intuition

* ($\Omega$): All possible realities
* ($\mathcal{F}$): Events you are allowed to ask probability about
* ($P$): The rule assigning likelihoods

---


## 4. Define Random Variable

Let
$$
X: (\Omega , \mathcal{F})  \to (S,  \mathcal{S}) 

\implies X(\omega) = s , \forall \omega \in \Omega , s \in S
$$
is a Random variable if
$$
X^{-1}(A) \in \mathcal{F} \quad \forall A \in \mathcal{S}
$$
### Example 1 : Discrete Case

$\Omega =\{HH, HT, TH, TT\} \\
\mathcal{F} = \{\emptyset, \{HH, HT\}, \{TH, TT\},\Omega\}$ 

Let: 
$$ X(\omega) =
\begin{cases}
1 & \text{if } \omega = \text{Head on first} \\
0 & \text{otherwise}
\end{cases}
$$


| Outcome | X(ω) |
|--------|------|
| HH     | 1    |
| HT     | 1    |
| TH     | 0    |
| TT     | 0    |



### Check Measurability

Take:
$$
A = \{1\}
$$

Then:
$$
X^{-1}(A) = \{HH, HT\} \in \mathcal{F}
$$

✅ Measurable

Let: 
$$ Y(\omega) =
\begin{cases}
1 & \text{if } \omega = \text{Head on both trial} \\
0 & \text{otherwise}
\end{cases}
$$

| Outcome | Y(ω) |
|--------|------|
| HH     | 1    |
| HT     | 0    |
| TH     | 0    |
| TT     | 0    |


Take:
$$
B = \{1\}
$$

Then:
$$
Y^{-1}(B) = \{HH\} \notin \mathcal{F}
$$

✅  Not Measurable
---

### Example 2: Continuous Case

### Setup

- $\Omega = \mathbb{R} $
- $\mathcal{F} = \mathcal{B}(\mathbb{R}) $ (Borel σ-algebra)

---

### Define Random Variable

$$
X(\omega) = \omega^2
$$


### Check Measurability

Take:
$$
A = [0,1]
$$

Then:
$$
X^{-1}(A) = [-1,1]
$$

Since:
$$
[-1,1] \in \mathcal{B}(\mathbb{R})
$$

✅ Measurable

---

## 5. Pushforward Measure (Distribution)

A random variable induces a new measure on the target space:

$$
\mu_X(A) = \mathbb{P}(X \in A) = \mathbb{P}(X^{-1}(A))
$$

---

### Example (from coin toss)

$$
\mu_X(0) = \frac{1}{4}, \quad
\mu_X(1) = \frac{1}{2}, \quad
\mu_X(2) = \frac{1}{4}
$$

---

### Key Relationships

- Event space $\mathcal{F} $: defines measurable events  
- Random variable $ X $: connects Ω to another space  
- Pushforward measure $ \mu_X $: gives distribution  

---

### Summary

- A **measurable space** defines which subsets are valid events  
- A **measure space** assigns probabilities to those events  
- A **random variable** is a measurable mapping:
  $
  X : (\Omega, \mathcal{F}) \to (S, \mathcal{S})
  $
- It induces a **distribution** on \( S \) via:
  $
  \mu_X = \mathbb{P} \circ X^{-1}
  $

---

### One-Line Intuition

> The event space defines what can be measured, and the random variable translates those measurable events into a structured space where probability becomes easier to analyze.

# From Measure-Theoretic Probability to Distributions

## 1. Big Picture

We start with a **probability space**:

$$
(\Omega, \mathcal{F}, \mathbb{P}) \xrightarrow{X} (S, \mathcal{S})
$$

which gives the pushforward(distribution):
$$ \mu_X (A) = P(X \in A)$$

This $\mu_X$ is what we plot

---

## 2. Discrete Example (Coin Toss -> Binomial Distribution)

### Setup

- Toss 3 coins
- $\Omega = \{HHH, HHT, HTH, THH, HTT, THT, TTH, TTT\}$
- $\mathcal{F} = 2^{\Omega}$
- $X(\omega) = $ number of heads

|   s    |$\mu_X =P(X^{-1}(s))$|
|--------|-------|
|  0     |  1/8   |
|  1     |  3/8    |
|  2     |  3/8    |
|  3     |  1/8    |

```python
import matplotlib.pyplot as plt

# Number of heads
s = [0, 1, 2, 3]

# Corresponding probabilities for 3 fair coin tosses
p = [1/8, 3/8, 3/8, 1/8]

plt.figure(figsize=(6,4))
plt.bar(s, p)

plt.xticks(s)
plt.xlabel("Number of Heads (s)")
plt.ylabel("Probability P(X = s)")
plt.title("Binomial Distribution from 3 Coin Tosses")

plt.show()
```

## 🔹 3. Continuous Example: Gaussian Distribution

### Definition:
$$
X \sim \mathcal{N}(\mu, \sigma^2)
$$

### Density:
$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \, e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

### Interpretation:
This is the density of the pushforward measure.

Instead of probabilities at points, we have:
$$
P(a \le X \le b) = \int_a^b f(x)\,dx
$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 100)
mu, sigma = 0, 1

y = (1/(np.sqrt(2*np.pi)*sigma)) * np.exp(-(x-mu)**2/(2*sigma**2))

plt.plot(x, y)
plt.title("Gaussian Distribution")
plt.xlabel("x")
plt.ylabel("density")
plt.show()
```

```python
import numpy as np

from scipy.stats import norm, expon, beta

# x-axis
x = np.linspace(-5, 5, 1000)

# Define densities
pdf_normal = norm.pdf(x, loc=0, scale=1)        # Standard normal
pdf_normal2 = norm.pdf(x, loc=2, scale=0.7)     # shifted normal

# Exponential needs positive domain
x_pos = np.linspace(0, 5, 1000)
pdf_expon = expon.pdf(x_pos, scale=1)

# Beta distribution on [0,1]
x_beta = np.linspace(0, 1, 1000)
pdf_beta = beta.pdf(x_beta, a=2, b=5)

# Plot
plt.figure(figsize=(8,5))

plt.plot(x, pdf_normal, label="N(0,1)", color="blue")
plt.plot(x, pdf_normal2, label="N(2,0.7)", color="red")

plt.plot(x_pos, pdf_expon, label="Exponential(1)", color="green")

plt.plot(x_beta, pdf_beta, label="Beta(2,5)", color="purple")

plt.title("Continuous Probability Density Functions")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid(True)

plt.show()
```

## 📘 Joint Distribution

### 🔷 Definition

The **joint distribution** describes the probability of two random variables occurring together.

$$
P(X, Y)
$$

For continuous variables, it is written as:

$$
f_{X,Y}(x,y)
$$

---

### 🔷 Example: 2D Gaussian Joint Distribution

Below is a Python example to visualize a joint distribution.



```python


from scipy.stats import multivariate_normal

# Define mean and covariance
mean = [0, 0.1]
cov = [[1, 0.8],
       [0.8, 1]]

# Create distribution
rv = multivariate_normal(mean, cov)

# Create grid
x = np.linspace(-3, 3, 200)
y = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

# Evaluate joint PDF
Z = rv.pdf(pos)

# Plot
plt.figure(figsize=(6,5))
plt.contourf(X, Y, Z, levels=50, cmap="viridis")
plt.colorbar(label="p(x,y)")
plt.title("Joint Distribution p(X, Y)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## 📘 Marginal Distribution
### 🔷 Definition

The **Marginal distribution** are obtained by integrating out variables.

$$P(x) = \int P(x,y) dy , \quad P(y) = \int P(x,y) dx $$

```python
px = np.trapezoid(Z, y, axis=0)
py = np.trapezoid(Z, x, axis=1)

plt.figure(figsize=(6,4))
plt.plot(x, px, label="p(X)", color="blue")
plt.plot(y, py, label="p(Y)", color="red")
plt.title("Marginal Distributions")
plt.legend()
plt.grid()
plt.show()
```

## 📘 Conditional Distribution
### 🔷 Definition



$$p(x|y) = \frac{p(x,y)}{p(y)}$$

```python
y_index = np.argmin(np.abs(y - 1))

joint_slice = Z[y_index, :]
px_given_y = joint_slice / np.trapezoid(joint_slice, x)

plt.figure(figsize=(6,4))
plt.plot(x, px_given_y, label="p(X | Y=1)", color="green")
plt.title("Conditional Distribution")
plt.legend()
plt.grid()
plt.show()
```

## 📘 Bayes' Theorem

$$
p(x, y) = p(x|y)\, p(y) = p(y|x)\, p(x)
$$

$$
p(y|x) = \frac{p(x|y)\, p(y)}{p(x)}
$$

---

### 🔷 Interpretation


$$
\underbrace{p(y|x)}_{\text{Posterior}} 
= 
\frac{
\overbrace{p(x|y)}^{\text{Likelihood}} \cdot 
\overbrace{p(y)}^{\text{Prior}}
}{
\underbrace{p(x)}_{\text{Evidence}}
}
$$

---


```python
x0_index = np.argmin(np.abs(x - 0.5))

px_val = px[x0_index]

py_given_x = Z[:, x0_index]
py_given_x = py_given_x / np.trapezoid(py_given_x, y)

bayes_recon = py_given_x * px_val
bayes_recon = bayes_recon / np.trapezoid(bayes_recon, y)

plt.figure(figsize=(6,4))
plt.plot(y, py_given_x, label="p(Y | X=0.5)", color="purple")
plt.title("Bayes Theorem Interpretation")
plt.legend()
plt.grid()
plt.show()
```

# 📘 Uncertainty
## 🔷 Definition
In the context of data science, engineering, and physics, uncertainty is the quantifiable doubt about the validity of a measurement or a prediction. It is the acknowledgement that our models and sensors are never perfect reflections of reality.
## Types of Uncertainty

### 1. Aleatoric Uncertainty (Irreducible Error)

Aleatoric uncertainty arises from the inherent randomness in the system or data. It represents noise that cannot be eliminated, even with infinite data or a perfect model.

It is commonly modeled as:

- y_obs = y + $ \epsilon$ (Additive) 
- y_obs = y(1+ $ \epsilon$ ) (Multiplicative noise)

where:
- y_obs = observed data  
- y = true (noise-free) value  
- $ \epsilon$ ~ N(0, σ²) = random noise  

### Key properties:
- Inherent in the data-generating process  
- Cannot be reduced by collecting more data  
- Must be modeled probabilistically  

---

### 2. Epistemic Uncertainty (Reducible Error)

Epistemic uncertainty arises due to lack of knowledge about the system, such as unknown parameters or imperfect models.

It is associated with uncertainty in model structure or parameters:

y = f(x; θ)

where:
- θ = unknown parameters  
- f = model approximation  

### Key properties:
- Due to incomplete knowledge or modeling assumptions  
- Can be reduced by:
  - collecting more data  
  - improving the model  
  - better parameter estimation  
- Includes model uncertainty and parameter uncertainty
